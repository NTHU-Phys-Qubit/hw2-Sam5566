# 1D Ising model + Transfer matrix

* Construct the transfer matrix at temperature $T$ with external field $B$
* Find the eivenvalues of the transfer matrix and compare to the exact results.
* Evaluate the partition function $Z$ at different system size $N$ and $T$, $B$. Compare to the exact results.
* For a fixed $N$, $T$, $B$, evaluate $\langle S_0 S_j\rangle$ where $j=0,1,\cdots N$. 
* Plot $\langle S_0 S_j\rangle$ up to $j=N/2$ and show that it decays exponentially.
  * Please try several $N$, $T$, $B$. What is the largest $N$ you can go? What is the lowest $T$ you can go?

```python
# CONSTANT

## boltzmann factor
k = 1 #1.38 * 10**23

## coupling constant
J = 1

## impurity matrix
I =  np.array([[1, 0], [0, -1]])
```

## Definition of (1) transfer matrix (2) eigenvalue (3) partition function (4) correlation function
 the class of transfer matrix contain these value

```python
# Property of transfer matrix
class TM: 
    def __init__(self, T, B):
        beta = 1/T/k
        # transfer matrix
        self.P = np.zeros((2,2), dtype = float) 
        self.P[0][0] = np.exp(beta * (J + B))
        self.P[0][1] = np.exp(-beta * J)
        self.P[1][0] = np.exp(-beta * J)
        self.P[1][1] = np.exp(beta * (J - B))
        # temperature
        self.T = T 
        # magnitic field
        self.B = B
        
    # lambda +
    def eign(self):
        return np.linalg.eig(self.P) 
    
    # partition function
    def Z(self, N):
        return np.trace(np.linalg.matrix_power(self.P, N))
    
    def SS(self, order, N):
        j = int(order)
        correlation_matrix = (I @ np.linalg.matrix_power(self.P, j) @ I @ np.linalg.matrix_power(self.P, N - j))
        correlation = np.trace(correlation_matrix) / np.trace(np.linalg.matrix_power(self.P, N))
        return correlation
```

## exact eigenvalues

```python
# the numeral calculation of eigenvalues & eignvectors (for the comparison)
def eign(T, B):
    lb_1 = np.exp(J/(k * T))*np.cosh(B/(k * T)) + np.sqrt(np.exp(-2*J/(k * T)) + np.exp(2*J/(k * T))*np.sinh(B/(k * T))**2)
    lb_2 = np.exp(J/(k * T))*np.cosh(B/(k * T)) - np.sqrt(np.exp(-2*J/(k * T)) + np.exp(2*J/(k * T))*np.sinh(B/(k * T))**2)
    return lb_1, lb_2  
```

### comparison between two kinds of eigenvalues

```python
tem = 1
mag = 0.1

eee = TM(tem, mag)

print ("eigenvalue is", eee.eign()[0] )
print ("exact eigenvalueis ", eign(tem, mag))
```

### comparison between two kinds of partition functions

```python
size = 10
print ("for size is 10,")
print ("exact partition is", eee.Z(size))
print ("thermodynamic limit is", eee.eign()[0][0]**size)

size = 100
print ("for size is 100,")
print ("exact partition is", eee.Z(size))
print ("thermodynamic limit is", eee.eign()[0][0]**size)
```

## The correlation functions of different size

```python
def SS_plot(N, T, B):
    a = np.zeros(N)
    M = TM(T, B)
    for i in range(N):
        a[i] = M.SS(i, N)
    return a
```

## Figure of correlation functions up to size equal to half of N

### different temperautures

```python
SS_128_T_100 = SS_plot(128, 1, 0)
SS_128_T_075 = SS_plot(128, 0.75, 0)
SS_128_T_025 = SS_plot(128, 0.25, 0)
```

```python
time = np.linspace(0, np.size(SS_128_T_100) - 1, np.size(SS_128_T_100))
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(time[0:64], SS_128_T_100[0:64], color='blue', linewidth=2.2, linestyle='-', label = 'T = 1')
ax.plot(time[0:64], SS_128_T_075[0:64], color='red', linewidth=2.2, linestyle='-', label = 'T = 0.75')
ax.plot(time[0:64], SS_128_T_025[0:64], color='green', linewidth=2.2, linestyle='-', label = 'T = 0.25')

ax.legend()
ax.set_xlabel('n',fontsize=15)
ax.set_ylabel('correlation',fontsize=15)
ax.set_title('site verse correlation',fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid()
```
